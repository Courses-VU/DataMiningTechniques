```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics 
from statistics import mean
import copy
```

```python
#read file
data=pd.read_csv("../Data_mining/data/dataset_mood_smartphone.csv")
data
```

### Preprocessing 

```python
prev_day="02-26"
new_data=pd.DataFrame(columns=['id', 'time', 'var',"value"])
temp_list=[]
for index, row in data.iterrows():
    
    date=row["time"].split(" ")[0]
    date=date.split("-")[1]+"-"+date.split("-")[2]
    if date != prev_day:
        new_data.loc[index]=[data.loc[index-1]["id"], prev_day , data.loc[index-1]["variable"], statistics.mean(temp_list)]
        temp_list=[row["value"]]
    else:
        temp_list.append(row["value"])
    prev_day=date

        
```

```python
new_data
```

```python
temps = pd.DataFrame(new_data.values)
dataframe = pd.concat([temps.shift(3), temps.shift(2), temps.shift(1), temps], axis=1)
dataframe.columns = ["id1","time1", "var1","val1","id2","time2", "var2","val2","id3","time3", "var3","val3","id4","time4", "var4","val4"]
dataframe.loc[3:].head()
```

```python
data=dataframe.loc[3:]
df= dataframe.loc[3:]
df['check_var'] = np.where((df['var1'] == df['var2']) & (df['var1'] == df['var3']) & (df['var1'] == df['var4'])
                     , 1 , 0)
df['check_id'] = np.where((df['id1'] == df['id2']) & (df['id1'] == df['id3']) & (df['id1'] == df['id4'])
                     , 1 , 0)
#df['check_date']= np.where((df["time1"].split("-")[0] == df["time2"].split("-")[0]) & 
#                           (df["time1"].split("-")[0] == df["time3"].split("-")[0]) &
#                           (df["time1"].split("-")[0] == df["time4"].split("-")[0]), 1, 0)
df
```

```python
print(df.shape)
#removing rows with different ID and variable
for i, row in df.iterrows():
    if row["check_id"]==0 or row["check_var"]==0:
        data=data.drop(i)
    
data=data.drop(['var2',"var3","var4","id2","id3","id4"], axis=1) 
print(data.shape)
data.head()
```

```python
data=data[["id1","var1","time1","val1","time2","val2","time3","val3","time4","val4"]].reset_index()
#data.columns = ['a', 'b']
data['time1'] = pd.to_datetime(data['time1'],format="%m-%d")
data['time2'] = pd.to_datetime(data['time2'],format="%m-%d")
data['time3'] = pd.to_datetime(data['time3'],format="%m-%d")
data['time4'] = pd.to_datetime(data['time4'],format="%m-%d")

data.head()
```

```python
data = data.drop(columns=["index"])
```

```python
print(data["time1"].iloc[0,])
print(data["time2"].iloc[0,])
delta2=data["time1"].iloc[0,] - data["time2"].iloc[0,]
delta2
```

```python
#check and remove rows with not consecutives days
data_iter=data
for i, row in data_iter.iterrows():
    #print(i)
    delta1=data_iter["time1"].iloc[i,] - data_iter["time2"].iloc[i,]
    delta2=data_iter["time2"].iloc[i,] - data_iter["time3"].iloc[i,]
    delta3=data_iter["time3"].iloc[i,] - data_iter["time4"].iloc[i,]
    delta1=delta1.days
    delta2=delta2.days
    delta3=delta3.days
    if delta1 != -1 or delta2 != -1 or delta3 != -1:
        #print(delta1, delta2, delta3)
        #print("ops")
        data = data.drop(i)
```

```python
data=data.reset_index()
data = data.drop(columns=["index"])
#data=data.drop(columns="level_0")
data.head()
```

```python
data["val1"] = data["val1"].apply(pd.to_numeric, errors='coerce')
data["val2"] = data["val2"].apply(pd.to_numeric, errors='coerce')
data["val3"] = data["val3"].apply(pd.to_numeric, errors='coerce')
data["val4"] = data["val4"].apply(pd.to_numeric, errors='coerce')
#drop dates
data=data.drop(['time1', 'time2',"time3"], axis=1)
#table = pd.pivot_table(data, values=["val1","val2","val3"], index=['id1', 'time4'], columns=['var1'])
#table["val1","mood"]

#add column with the average 
data['avg'] = data[['val1', 'val2','val3']].mean(axis=1)
```

```python
slopes=[]
for i, row in data.iterrows():
    a,b = np.polyfit([1,2,3],[row["val1"],row["val2"],row["val3"]],1)
    slopes.append(a)
data["slopes"]=slopes

data
```
