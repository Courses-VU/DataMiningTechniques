```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics 
import copy
```

```python
#read file
data=pd.read_csv("../Data_mining/data/dataset_mood_smartphone.csv")
data
```

### Preprocessing 

```python
prev_day="02-26"
new_data=pd.DataFrame(columns=['id', 'time', 'var',"value"])
temp_list=[]
for index, row in data.iterrows():
    
    date=row["time"].split(" ")[0]
    date=date.split("-")[1]+"-"+date.split("-")[2]
    if date != prev_day:
        new_data.loc[index]=[data.loc[index-1]["id"], prev_day , data.loc[index-1]["variable"], statistics.mean(temp_list)]
        temp_list=[row["value"]]
    else:
        temp_list.append(row["value"])
    prev_day=date

        
```
